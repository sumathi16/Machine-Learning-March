#### Removing weired spaces

```python
data = "  hello This    is Sumathi. Isn't it beautiful?\t  "
```

```python
def remove_space(text):
    text = text.strip()
    print(text)
    text = text.split()
    print(text)
    return " ".join(text)
```

```python
from spellchecker import SpellChecker

spell = SpellChecker()
def correct_spellings(text):
    corrected_text = []
    misspelled_words = spell.unknown(text.split())
    for word in text.split():
        if word in misspelled_words:
            corrected_text.append(spell.correction(word))
        else:
            corrected_text.append(word)
    return " ".join(corrected_text)
        
text = "everone"
correct_spellings(text)
```

```python
correct_spellings("hallo how are you?")
```

#### Contraction

```python
contraction = {
 "'cause": 'because',
 ',cause': 'because',
 ';cause': 'because',
 "ain't": 'am not',
 'ain,t': 'am not',
 'ain;t': 'am not',
 'ain´t': 'am not',
 'ain’t': 'am not',
 "aren't": 'are not',
 'aren,t': 'are not',
 'aren;t': 'are not',
 'aren´t': 'are not',
 'aren’t': 'are not'
}
```

```python
def mapping_replacer(x, dic):
    for word in dic.keys():
        if " " + word + " " in x:
            x = x.replace(" " + word + " ", " " + dic[word] + " ")
    return x
```

```python
mapping_replacer('hello ;cause ',contraction)
```

#### Emoji Handling

```python
# Reference : https://gist.github.com/slowkow/7a7f61f495e3dbb7e3d767f97bd7304b
import re
def remove_emoji(string):
    emoji_pattern = re.compile("["
                           u"\U0001F600-\U0001F64F"  # emoticons
                           u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                           u"\U0001F680-\U0001F6FF"  # transport & map symbols
                           u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                           u"\U00002702-\U000027B0"
                           u"\U000024C2-\U0001F251"
                           "]+", flags=re.UNICODE)
    return emoji_pattern.sub(r'', string)

remove_emoji("game is on 🔥🔥")
```

### Traditional handling of text data
- Hashing of words
- Count vectorization
- TF-IDF
- SVD

- TF(t) = Number of times a term t appears in a document / Total number of terms in the document
- IDF(t) = LOG( Total number of documents/Number of documents with term t in it )
- TF-IDF(t) = TF(t) * IDF(t)

```python
from sklearn.feature_extraction.text import TfidfVectorizer
corpus = [
    'This is the first document.',
    'This document is the second document.',
    'And this is the third one.',
    'Is this the first document?',
]
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(corpus)
print(vectorizer.get_feature_names())

print(X.shape)
print(X.toarray())
```
