```python
feedback = ["Okay Its good",
            "Offline sessions were better",
            "Because of network issue its not good.",
            "It's bad",
            "May be we can do better than this",
            "Less practical",
            "More theoritical oriented",
            "Not explaing in depth.",
            "Good", 
            "Excellent!", 
            "The sesions were good I am satisfied with content.",
            "I am satisfied with class."]
```

```python
feedback[0].split()
```

```python
words = []
punctuations = ["'",'.','!']
sentences = []
for sen in feedback:
    sen = sen.lower()
    for c in punctuations:
        sen = sen.replace(c,'')
    sen_words = sen.split()
    sentences.append(sen_words)
    words.extend(sen_words)
```

```python
words
```

```python
sentences
```

```python
unique_words = list(set(words))
unique_words
```

```python
word_count =[]
for each in sentences:
    wc = []
    for word in unique_words:
        wc.append(each.count(word))
    word_count.append(wc)
```

```python
print(word_count)
```

```python
import pandas as pd
data = pd.DataFrame(word_count,columns=unique_words)
```

```python
import pandas as pd
data = pd.read_csv("data/smsspamcollection/SMSSpamCollection",sep ="\t",names=["spam","message"])
data.head()
```

```python
data['message'][:5]
```

```python
print(data['message'][1])
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vect = CountVectorizer()
out = vect.fit_transform(data['message'])
x = pd.DataFrame(out.toarray(),columns = vect.get_feature_names())
```

```python
x
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vect = CountVectorizer(stop_words='english')
out = vect.fit_transform(data['message'])
x = pd.DataFrame(out.toarray(),columns = vect.get_feature_names())
```

```python
x
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vect = CountVectorizer(stop_words='english',lowercase=True)
out = vect.fit_transform(data['message'])
x = pd.DataFrame(out.toarray(),columns = vect.get_feature_names())
```

```python
y = data['spam'].apply(lambda x:1 if x=='spam' else 0)
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y)
```

```python
from sklearn.naive_bayes import BernoulliNB
BNB = BernoulliNB()
BNB.fit(x_train,y_train)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(BNB.predict(x_test),y_test)
```
